```python
#Feature Selection
features=recommendation[['Movie_Genre','Movie_Tagline','Movie_Overview','Movie_Keywords','Movie_Cast','Movie_Director',]].fillna('')
features.head()
```

```python
X=features['Movie_Genre']+' '+features['Movie_Keywords']+' '+features['Movie_Tagline']+' '+features['Movie_Cast']+' '+features['Movie_Director']
```

```python
print(X.shape)
X
```

```python

#Text feature conversion to tokens

#In order to process natural language, the text must be represented as a numerical feature.
#The process of transforming text into a numerical feature is called text vectorization
from sklearn.feature_extraction.text import TfidfVectorizer
#Tfid(Term Frequency Inverse Document Frequency) is a text vectorization algorithm,
#which helps in transforming text into vector
tfid=TfidfVectorizer()
```

```python
X=tfid.fit_transform(X)
```

```python
X.shape
```

```python
X
print(X)
```

```python
#Getting Similarity Score By Using Cosine similarity

#similarity measure refers to distance with dimensions representing features of the data object, 
#in a dataset. If this distance is less, there will be a high degree of similarity and vice versa 
from sklearn.metrics.pairwise import cosine_similarity
#In cosine similarity, data objects in a dataset are treated as a vector. 
#The formula to find the cosine similarity between two vectors is Cos(x, y) = x . y / ||x|| * ||y||
similarity_score=cosine_similarity(X)
```

```python
similarity_score
```

```python
#Get user input for movie name and validating for closest spelling

movie_name=input("Enter your favourite Movie Name: ")
```

```python
movies_title_list=recommendation['Movie_Title'].tolist()
```

```python
import difflib as dl
#difflib module contains several functions and classes that allow user to compare sets of data
movie_recommendtion=dl.get_close_matches(movie_name,movies_title_list)
print(movie_recommendtion)
```

```python
index_close_match=recommendation[recommendation.Movie_Title==movie_recommendtion[0]]['Movie_ID'].values[0]
print(index_close_match)
```

```python
#Getting a list of similar movies
recommendation_score=list(enumerate(similarity_score[index_close_match]))
print(recommendation_score)
```

```python
#Get all Movies sort based on Recommendation score
sorted_similar_movies=sorted(recommendation_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_movies)
```

```python
#Print the name of similar movie based on the index

print("Top 15 movies recommended for you:")
k=1

for movie in sorted_similar_movies:
  index=movie[0]
  title_from_index=recommendation[recommendation.index==index]['Movie_Title'].values[0]
  if(k<16):
    print(k,' ',title_from_index)
    
  k=k+1
```

#Top 10 Movie recommendation

```python
movie_name=input("Enter your favourite Movie Name: ")
movies_title_list=recommendation['Movie_Title'].tolist()

import difflib as dl
movie_recommendtion=dl.get_close_matches(movie_name,movies_title_list)
index_close_match=recommendation[recommendation.Movie_Title==movie_recommendtion[0]]['Movie_ID'].values[0]
recommendation_score=list(enumerate(similarity_score[index_close_match]))
sorted_similar_movies=sorted(recommendation_score,key=lambda x:x[1],reverse=True)

#Print the name of similar movie based on the index

print("Top 10 movies recommended for you:")
k=1

for movie in sorted_similar_movies:
  index=movie[0]
  title_from_index=recommendation[recommendation.index==index]['Movie_Title'].values[0]
  if(k<11):
    print(k,' ',title_from_index)
  k=k+1
```
